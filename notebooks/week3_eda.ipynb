```python
import os
import pandas as pd
import json

# 1. Проверяем текущую папку
print(f"Мы тут: {os.getcwd()}")

# 2. Ищем файлы в data/raw
raw_dir = "../data/raw"
if os.path.exists(raw_dir):
    files = os.listdir(raw_dir)
    print(f"Найденные файлы: {files}")
else:
    print("Папка data/raw не найдена! Проверь структуру.")
```

```python
# Читаем JSON файл
file_path = "../data/raw/holidays_DE_2025_20260318_1157.json"

with open(file_path, 'r', encoding='utf-8') as f:
    raw_data = json.load(f)

# Создаем таблицу (DataFrame)
df = pd.DataFrame(raw_data)

# Показываем первые 5 строк и информацию о колонках
print("Размер таблицы:", df.shape)
df.head()
```

```python
# 1. Преобразуем колонку 'date' из текста в формат даты
df['date'] = pd.to_datetime(df['date'])

# 2. Убираем лишние скобки в колонке 'types' (было ['Public'], станет Public)
df['types'] = df['types'].apply(lambda x: x[0] if isinstance(x, list) else x)

# 3. Отбираем только полезные колонки и переименовываем их
# Уберем 'fixed', 'global', 'counties' и 'launchYear', так как они почти пустые или не нужны
df_normalized = df[['date', 'localName', 'name', 'types']].copy()
df_normalized.columns = ['holiday_date', 'local_name', 'eng_name', 'category']

print("Данные очищены! Вот как теперь выглядит таблица:")
df_normalized.head()
```

```python
import os
from datetime import datetime

# Создаем папку для нормализованных данных
output_folder = "../data/normalized/variant_20"
os.makedirs(output_folder, exist_ok=True)

# Генерируем имя файла с датой и временем
now = datetime.now().strftime("%Y-%m-%d_%H-%M-%S")
file_name = f"{now}_holidays_germany.csv"
output_path = os.path.join(output_folder, file_name)

# Сохраняем без лишнего столбца индексов
df_normalized.to_csv(output_path, index=False)

print(f"Файл успешно сохранен по пути:\n{output_path}")
```
