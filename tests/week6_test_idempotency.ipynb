```python
import pandas as pd
import os

new_data = pd.DataFrame({"id": [1, 2], "v": [10, 20]})

if os.path.exists("out.csv"):
    old_data = pd.read_csv("out.csv")
    df = pd.concat([old_data, new_data])
    df = df.drop_duplicates(subset=["id"])
else:
    df = new_data

df.to_csv("out.csv", index=False)

print(f"В файле сейчас строк: {len(df)}")
```
